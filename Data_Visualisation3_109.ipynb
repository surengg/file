```python
import matplotlib.pyplot as plt 
import numpy as np

x = np.array(["A", "B", "C", "D"]) 
y = np.array([5, 7, 2, 15])

plt.bar(x,y)

plt.show()
```

```python
plt.bar([0.5,2.5,4.5,6.5,8.5],[50,40,70,80,20],
        label="BMW",width=1) 

plt.bar([1.5,3.5,5.5,7.5,9.5], [80,20,20,50,60], 
        label="Audi", color='c', width=1)

plt.legend()

plt.xlabel('Days')

plt.ylabel('Distance (kms)')

plt.title('Information')

plt.show()
```

```python
plt.barh(x,y)
plt.show()
```

```python
plt.bar(x,y, color='red')
plt.show()
```

```python
plt.bar(x,y, color='g', width=0.1)
plt.show()
```

```python
x = np.random.normal(300, 20, 250)
plt.hist(x)
plt.show()
```

```python
y = np.array([40, 30, 20, 10])
plt.pie(y)
plt.show()
```

```python
y = np.array([35, 25, 25, 15])

mylabels = ["MPMC", "PPML", "ContolSystem", "PSA"]

plt.pie(y, labels = mylabels)

plt.show()

```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["MPMC", "PPML", "ContolSystem", "PSA"]
myexplode = [0, 0, 0.3, 0]
plt.pie(y,labels=mylabels,explode=myexplode)
plt.show()
```
