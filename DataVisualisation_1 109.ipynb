```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([0,6])
y = np.array([0,250])
print(plt.plot(x,y))
print(plt.show())
```

```python
x = np.array([0,5])
y = np.array([0,25])
print(plt.plot(x,y,marker='p'))
print(plt.show())
```

```python
x = np.array([1,2,6,8])
y = np.array([3,8,1,10])
print(plt.plot(x,y))
print(plt.show())
```
