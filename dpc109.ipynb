```python
import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.randn(5,3),index=['a','c','e','f','h'],columns=['one','two','three'])
print(df)
df = df.reindex(['a','b','c','d','e','f','g','h'])
print('Reindex',df)
print('After filled ',df.fillna(0))
print('Dropped Values',df.dropna())
print('Null check',df['one'].isnull())
print('Replacing NaN values with forward values',df.fillna(method='pad'))
print('Replacing with backward values',df.fillna(method='bfill'))

df = pd.DataFrame({'one':[10,20,30,40],'two':[1000,0,30,90]})
print("Another Dataframe",df)
print('Replaced Values',df.replace({1000:20,90:2000,}))
```
